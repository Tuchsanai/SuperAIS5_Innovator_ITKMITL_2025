```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def process_image(image: np.ndarray, prompt: str) -> np.ndarray:
    # Copy image to avoid modifying original

    # random rotate image
    angle = np.random.randint(0, 360)
    h, w = image.shape[:2]
    center = (w // 2, h // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image, roation_matrix, (w, h))


    image_with_text = rotated_image.copy()
    
    # Set font parameters
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    thickness = 2
    color = (255, 0, 0)  # Red text in RGB (will look blue if not converted to RGB)

    # Put the text at the top-left corner
    cv2.putText(image_with_text, prompt, (10, 30), font, font_scale, color, thickness, cv2.LINE_AA)

    return image_with_text


# Display image using matplotlib
def display_image(image_np: np.ndarray, title="Processed Image"):
    plt.imshow(image_np)
    plt.axis("off")
    plt.title(title)
    plt.show()
```

```python

# Example usage
file_path = 'cat1.png'  # Replace with your actual image file
prompt = " Hello this is my cat"  # Or change prompt

# Load image using OpenCV and convert BGR to RGB
image_bgr = cv2.imread(file_path)
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

display_image(image_rgb, title="Original Image")
```

```python

# Process and display
processed_image = process_image(image_rgb, prompt)
display_image(processed_image)
```
