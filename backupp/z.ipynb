```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Occupation Distribution (Bar Plot with Counts)
plt.figure(figsize=(12, 8))

# Create countplot
ax = sns.countplot(
    y='Occupation_Category', 
    data=df, 
    order=df['Occupation_Category'].value_counts().index
)

# Annotate each bar with the count
for p in ax.patches:
    count = int(p.get_width())                  # get bar length
    x = p.get_width() + df.shape[0] * 0.005      # offset: tweak multiplier as needed
    y = p.get_y() + p.get_height() / 2           # vertical center of bar
    ax.text(
        x, y,                                  # position
        count,                                 # the text
        va='center',                           # vertical alignment
        fontsize=10
    )

plt.title('Distribution of Occupations')
plt.xlabel('Count')
plt.ylabel('Occupation Category')

plt.tight_layout()
plt.show()
```
